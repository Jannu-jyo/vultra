```python
import dlib
import cv2
import pyttsx3
import numpy as np
# Attempting imports for OLED
try:
    from luma.core.interface.serial import i2c
    from luma.core.render import canvas
    from luma.oled.device import ssd1306
    HAS_DISPLAY = True
except ImportError:
    HAS_DISPLAY = False

class VultraRobot:
    def __init__(self):
        # 1. Voice Setup
        self.engine = pyttsx3.init()
        
        # 2. Vision Setup (dlib)
        self.detector = dlib.get_frontal_face_detector()
        
        # 3. OLED Display Setup (Safety check for Windows)
        self.device = None
        if HAS_DISPLAY:
            try:
                self.serial = i2c(port=1, address=0x3C)
                self.device = ssd1306(self.serial)
            except Exception as e:
                print(f"OLED Hardware not found: {e}")

    def update_screen(self, status):
        """Only updates if physical device is present"""
        if self.device:
            with canvas(self.device) as draw:
                draw.rectangle(self.device.bounding_box, outline="white", fill="black")
                draw.text((30, 25), "VULTRA AI", fill="white")
                draw.text((20, 45), f"STATUS: {status}", fill="white")
        else:
            print(f"[OLED Simulator]: {status}")

    def run(self):
        cap = cv2.VideoCapture(0)
        self.update_screen("ONLINE")
        self.engine.say("VULTRA systems active.")
        self.engine.runAndWait()

        while True:
            ret, frame = cap.read()
            if not ret: break
            
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = self.detector(gray)
            
            for face in faces:
                x, y, w, h = (face.left(), face.top(), face.width(), face.height())
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                # Face Detection Event
                self.update_screen("HUMAN DETECTED")

            cv2.imshow('VULTRA Vision Hub', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'): break

        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    VultraRobot().run()
```

```python
import cv2
import face_recognition
import pyttsx3
import os
import numpy as np
import time

# --- 1. SETUP VOICE ---
engine = pyttsx3.init()
def vultra_speak(text):
    print(f"[VULTRA]: {text}")
    engine.say(text)
    engine.runAndWait()

# --- 2. LOAD YOUR PHOTO INTO MEMORY ---
known_face_encodings = []
known_face_names = []

vultra_speak("Loading my memory...")
faces_path = "faces/" 

if not os.path.exists(faces_path):
    print("Error: Please create a folder named 'faces' and put your photo in it!")
else:
    for filename in os.listdir(faces_path):
        if filename.endswith(".jpg"):
            img = face_recognition.load_image_file(f"{faces_path}{filename}")
            encoding = face_recognition.face_encodings(img)[0]
            known_face_encodings.append(encoding)
            known_face_names.append(filename.split(".")[0])

# --- 3. RECOGNITION ENGINE ---
def start_vultra_vision():
    # Fix: Try index 0, then 1 if it fails
    cap = cv2.VideoCapture(0)
    
    # CRITICAL FIX: Give camera 2 seconds to warm up
    print("Waking up camera...")
    time.sleep(2) 
    
    if not cap.isOpened():
        vultra_speak("I still cannot see. Check if your camera is covered.")
        return

    vultra_speak("Vision systems online. Hello everyone.")
    last_greeted = ""

    while True:
        ret, frame = cap.read()
        if not ret or frame is None: continue

        # Speed up recognition by scaling down the image
        small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
        rgb_small_frame = cv2.cvtColor(small_frame, cv2.COLOR_BGR2RGB)

        # Find faces
        face_locations = face_recognition.face_locations(rgb_small_frame)
        face_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)

        for face_encoding in face_encodings:
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "Unknown"

            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            if len(face_distances) > 0:
                best_match_index = np.argmin(face_distances)
                if matches[best_match_index]:
                    name = known_face_names[best_match_index]

            # VULTRA speaks the name
            if name != last_greeted:
                if name == "Unknown":
                    vultra_speak("I see a new person.")
                else:
                    vultra_speak(f"System identifies: {name}. Welcome back.")
                last_greeted = name

        cv2.imshow('VULTRA Vision Hub', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    start_vultra_vision()
```

```python
import cv2
import numpy as np
import speech_recognition as sr
import pyttsx3
import ollama
import time

# --- 1. THE VOICE ENGINE ---
def vultra_speak(text, emergency=False):
    print(f"\n[VULTRA]: {text}")
    try:
        engine = pyttsx3.init()
        # If it's an emergency, speak faster and louder
        rate = 200 if emergency else 160
        engine.setProperty('rate', rate)
        engine.say(text)
        engine.runAndWait()
        engine.stop()
    except Exception as e:
        print(f"Audio Error: {e}")

# --- 2. EMERGENCY TRIGGER FUNCTION ---
def trigger_fire_alert(frame):
    # Visual Alert on screen
    cv2.rectangle(frame, (0,0), (frame.shape[1], frame.shape[0]), (0,0,255), 25)
    cv2.putText(frame, "!!! FIRE ACCIDENT !!!", (70, 240), 
                cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,0,255), 4)
    vultra_speak("EMERGENCY ALERT! FIRE DETECTED IN THIS SECTOR. EVACUATE IMMEDIATELY!", emergency=True)

# --- 3. MAIN SYSTEM ENGINE ---
def start_vultra():
    # Use index 0 and DSHOW for stability on your laptop
    cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
    time.sleep(2.0) # Hardware warm-up (Point 3 of Reset)
    
    if not cap.isOpened():
        print("Hardware Error: Camera not found. Perform 4-Point Reset.")
        return

    r = sr.Recognizer()
    vultra_speak("VULTRA Online. Fire Guard is active in the background.")

    while True:
        ret, frame = cap.read()
        if not ret or frame is None:
            continue

        # --- STEP A: THE FIRE WATCH (Constant Monitoring) ---
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        lower_fire = np.array([0, 120, 70])
        upper_fire = np.array([25, 255, 255])
        mask = cv2.inRange(hsv, lower_fire, upper_fire)
        fire_pixels = cv2.countNonZero(mask)

        # --- STEP B: THE OCCUPATION CHECK ---
        # If fire color occupies a large area (15,000 pixels)
        if fire_pixels > 15000:
            trigger_fire_alert(frame)
            cv2.imshow("VULTRA Vision Hub", frame)
            cv2.waitKey(1)
            continue # Skip voice interaction during an accident

        # --- STEP C: NORMAL INTERACTION COMMANDS ---
        cv2.imshow("VULTRA Vision Hub", frame)
        
        try:
            with sr.Microphone() as source:
                r.adjust_for_ambient_noise(source, duration=0.3)
                audio = r.listen(source, timeout=0.1, phrase_time_limit=3)
                user_text = r.recognize_google(audio).lower()
                print(f"Heard: {user_text}")

                # New Commands:
                if "who created you" in user_text:
                    vultra_speak("I was created by Jyothi and the team at Ashoka Women's Engineering College.")
                
                elif "greet the judges" in user_text:
                    vultra_speak("Hello honorable judges. Welcome to the VULTRA project demonstration.")
                
                elif "system status" in user_text:
                    vultra_speak("All systems nominal. Vision sensors are scanning for thermal anomalies.")

                # Default AI Interaction (Ollama)
                elif "vultra" in user_text:
                    response = ollama.chat(model='llama3.2:1b', messages=[
                        {'role': 'user', 'content': f"One short sentence answer: {user_text}"}
                    ])
                    vultra_speak(response['message']['content'])
        except:
            pass

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    start_vultra()
```

```python
import cv2
import numpy as np
import speech_recognition as sr
import pyttsx3
import ollama
import threading
import time
import queue
import re
import requests
import random

# --- 1. GLOBAL STATE ---
speech_queue = queue.Queue()
is_vultra_talking = False

# --- 2. VOICE ENGINE ---
def speech_handler():
    global is_vultra_talking
    while True:
        text = speech_queue.get()
        if text is None: break
        is_vultra_talking = True
        try:
            engine = pyttsx3.init()
            engine.setProperty('rate', 145)
            print(f"\n[VULTRA]: {text}")
            engine.say(text)
            engine.runAndWait()
            engine.stop()
            del engine
        except: pass
        is_vultra_talking = False
        speech_queue.task_done()
        time.sleep(0.4)

def vultra_speak(text):
    if not text: return
    # Filter out LaTeX symbols for clear speech
    text = text.replace('^2', ' squared').replace('*', ' times ').replace('/', ' divided by ')
    lines = re.split(r'(?<=[.!?])\s+|(?=\d\.)', text)
    for line in lines:
        if line.strip(): speech_queue.put(line.strip())

threading.Thread(target=speech_handler, daemon=True).start()

# --- 3. SPECIAL MODULES (News, Games, Math) ---
def get_top_news():
    try:
        url = "https://saurav.tech/NewsAPI/top-headlines/category/technology/in.json"
        response = requests.get(url).json()
        articles = response['articles'][:3] 
        news_report = "Here are the top 3 technology headlines. "
        for i, article in enumerate(articles, 1):
            news_report += f"Headline {i}: {article['title']}. "
        return news_report
    except:
        return "I cannot reach the news servers, but I am ready for the expo."

def play_rps(user_choice_text):
    choices = {1: "Rock", 2: "Paper", 3: "Scissor"}
    vultra_choice = random.randint(1, 3)
    user_num = 0
    if "1" in user_choice_text or "rock" in user_choice_text: user_num = 1
    elif "2" in user_choice_text or "paper" in user_choice_text: user_num = 2
    elif "3" in user_choice_text or "scissor" in user_choice_text: user_num = 3
    
    if user_num == 0: return "Please say 1, 2, or 3."
    if user_num == vultra_choice: return f"Tie! We both chose {choices[vultra_choice]}."
    elif (user_num == 1 and vultra_choice == 3) or (user_num == 2 and vultra_choice == 1) or (user_num == 3 and vultra_choice == 2):
        return f"You win! {choices[user_num]} beats {choices[vultra_choice]}."
    else: return f"I win! {choices[vultra_choice]} beats {choices[user_num]}."

# --- 4. THE MASTER BRAIN ---
def brain_loop():
    r = sr.Recognizer()
    r.energy_threshold = 600
    
    # FORCED INTRODUCTION
    time.sleep(2)
    vultra_speak("Hello, Respected Judges. I am VULTRA, from Ashoka Women's Engineering College.")
    vultra_speak("I can now solve mathematical equations and provide complex formulas. How can I help?")

    while True:
        if not is_vultra_talking and speech_queue.empty():
            try:
                with sr.Microphone() as source:
                    print(">>> VULTRA IS LISTENING...")
                    r.adjust_for_ambient_noise(source, duration=0.8)
                    audio = r.listen(source, timeout=10, phrase_time_limit=8)
                
                user_text = r.recognize_google(audio).lower().strip()
                print(f"Heard: {user_text}")

                if any(name in user_text for name in ["vultra", "vuntra", "ultra"]):
                    
                    # 1. MATH & FORMULA LOGIC
                    if any(word in user_text for word in ["plus", "minus", "times", "divided", "formula", "equation", "square"]):
                        vultra_speak("Calculating the mathematical result.")
                        try:
                            response = ollama.chat(model='llama3.2:1b', messages=[
                                {'role': 'system', 'content': 'You are a Math Expert. Provide the formula or solve the equation in 1 very clear sentence for speech.'},
                                {'role': 'user', 'content': user_text}
                            ])
                            vultra_speak(response['message']['content'])
                        except:
                            vultra_speak("I am having trouble with my math processor.")

                    # 2. GAME LOGIC
                    elif "play" in user_text or "game" in user_text:
                        vultra_speak("Say 1 for Rock, 2 for Paper, or 3 for Scissor.")
                        with sr.Microphone() as game_source:
                            move_audio = r.listen(game_source, timeout=5)
                            move_text = r.recognize_google(move_audio).lower()
                            vultra_speak(play_rps(move_text))
                    
                    # 3. NEWS LOGIC
                    elif "news" in user_text:
                        vultra_speak(get_top_news())
                    
                    # 4. NORMAL AI ANSWER
                    else:
                        try:
                            response = ollama.chat(model='llama3.2:1b', messages=[
                                {'role': 'system', 'content': 'You are VULTRA. 1-sentence answer.'},
                                {'role': 'user', 'content': user_text}
                            ])
                            vultra_speak(response['message']['content'])
                        except:
                            vultra_speak("Standing by.")
                    
                    speech_queue.join()
                    vultra_speak("Next calculation?")

            except (sr.WaitTimeoutError, sr.UnknownValueError):
                continue 
            except Exception as e:
                print(f"Mic Resetting... Waiting 5s.")
                time.sleep(5)
        else:
            time.sleep(0.5)

# --- 5. DASHBOARD ---
def run_vultra():
    threading.Thread(target=brain_loop, daemon=True).start()
    cv2.namedWindow("VULTRA Dashboard")
    while True:
        ui = np.full((300, 600, 3), (25, 25, 25), dtype=np.uint8)
        status = "LISTENING" if not is_vultra_talking else "SPEAKING"
        cv2.putText(ui, f"VULTRA: {status}", (150, 150), 1, 2, (0, 255, 0), 2)
        cv2.imshow("VULTRA Dashboard", ui)
        if cv2.waitKey(1) & 0xFF == 27: break
    cv2.destroyAllWindows()

if __name__ == "__main__":
    run_vultra()
```
